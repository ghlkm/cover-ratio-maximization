```python
cardP=100
pdt_pset=sampleP(cardD, cardP)
def coverWho(user, sk,pdt, pdt_p, debug=True):
    covercnt=np.zeros(len(pdt_p))
    covered=[]
    for uid in range(len(user)):
        uT=user[uid].T
        for i in range(len(pdt_p)):
            pid=pdt_p[i]
            if np.dot(pdt[pid], uT)>sk[uid]:
                covered.append(uid)
                covercnt[i]+=1
    pdt_c=[]
    for i in range(len(covercnt)):
        if(covercnt[i]!=0):
            pdt_c.append(pdt_p[i])
            if debug:
                print(pdt_p[i], ":", covercnt[i], " # ", pdt[pdt_p[i]])
    return covered, pdt_c
def remove_covered(pdt, pdt_p, user, sk):
    covered, pdt_c=coverWho(user, sk, pdt, pdt_p)
    covered=set(covered)
    uncovered=[]
    for i in range(len(user)):
        if i not in covered:
            uncovered.append(i)
    return np.array([user[i] for i in uncovered]), np.array([sk[i] for i in uncovered]), pdt_c, covered
print("uni----")
user_uni1, sk_uni1, pdt_c_uni1, covered_uni1=remove_covered(pdt_uni, pdt_pset, user, sk_uni)
print("anti---")
user_anti1,sk_anti1, pdt_c_anti1, covered_anti1=remove_covered(pdt_anti, pdt_pset, user, sk_anti)
print("corr---")
user_corr1,sk_corr1, pdt_c_corr1, covered_corr1=remove_covered(pdt_corr, pdt_pset, user, sk_corr)
print(len(user_uni1), len(user_anti1), len(user_corr1))
plot_pdt_hs(pdt_uni, pdt_pset,pdt_c_uni1, user, sk_uni, covered_uni1)
plot_pdt_hs(pdt_anti, pdt_pset,pdt_c_anti1, user, sk_anti, covered_anti1)
plot_pdt_hs(pdt_corr, pdt_pset,pdt_c_corr1, user, sk_corr, covered_corr1)
```

```python
# # solve with SciPy
# from scipy.optimize import linprog
# c = [1, 1]
# A = [[1, 1], [1, -1]]
# b = [1, 0]
# x0_bounds = (0, 1)
# x1_bounds = (0, 1)
# res = linprog(c, A_eq=A, b_eq=b, \
#               bounds=(x0_bounds, x1_bounds),
#               options={"disp": False})
# print(res.success)
from scipy.optimize import linprog
def remove_unintersect(user, sk, row, B):
    intersect=[]
    target=[1, 1]
    x0_bounds = (0, 1)
    x1_bounds = (0, 1)
    for uid in range(len(user)):
        try:
            res=linprog(target, A_eq=[user[uid], row], b_eq=[sk[uid], B], 
                    bounds=(x0_bounds, x1_bounds), options={"disp": False})
            if res.success:
                intersect.append(uid)
        except:
            pass
    return intersect
constrain=np.array([1, 1])
```

```python
import numpy as np
import random as r
uniform_=r.uniform
gauss_=r.gauss
def uniform(a=0, b=1):
    return uniform_(a, b)
def gauss():
    ret=-1
    while ret<0 or ret>1:
        ret=gauss_(.5, 0.15)
    return ret

def dot6(nums):
    nums*=10**6
    nums=np.floor(nums)
    nums/=10**6
    return nums
def norm(data):
    for i in range(len(data[0])):
        dmin=min(data[:, i])
        dmax=max(data[:, i])
        dd=dmax-dmin
        data[:,i]-=dmin
        data[:, i]/=dd
    return data
def dot6_norm(data):
    data=dot6(data)
    data=norm(data)
    return data
```

```python

def sample_point_uni(d, b=1.0):
    left=b
    ret=[]
    for i in range(d-1):
        tmp=(1-(uniform(0, 1))**(1/(d-i-1)))
        ret.append(left*tmp)
        left-=ret[-1]
    ret.append(left)
    return ret
def gen_user_uni(n, d):
    ret =[sample_point_uni(d) for _ in range(n)]
    return ret
def gen_pdt_uni(n, d):
    ret=[]
    for i in range(n):
        ret.append([uniform() for _ in range(d)])
    return ret
def gen_pdt_anti(n, d):
    ret =[sample_point_uni(d) for _ in range(n)]
    return ret
def gen_pdt_corr(n, d):
    mean=[.5 for _ in range(d)]
    cov=[]
    for i in range(d):
        tmp=[]
        for j in range(d):
            if i==j:
                tmp.append(1)
            else:
                tmp.append(.95)
        cov.append(tmp)
    data = np.random.multivariate_normal(mean, cov, n)
    data=dot6_norm(data)
    return data
    
```

```python
import matplotlib.pyplot as plt
def plot_user(user):
    fig, ax=plt.subplots()
    user_a=np.array(user).T
    ax.scatter(user_a[0], user_a[1])
    ax.set(xlim=(0, 1), ylim=(0, 1))
    ax.set_aspect('equal', 'box')
    plt.show()
def plot_pdt(pdt):
    plot_user(pdt)
def plot_halfspace(r, c):
    fig, ax=plt.subplots()
    ax.set(xlim=(0, 1), ylim=(0, 1))
    ax.set_aspect('equal', 'box')
    for i in range(len(r)):
        row=r[i]
        if row[0]==0 or row[1]==0:
            continue
        ax.plot([c[i]/row[0], 0],[0, c[i]/row[1]])
    plt.show()
def plot_pdt_hs(pdt, pdt_p,pdt_c, r, c, covered=set()):
    fig, ax=plt.subplots()
    ax.set(xlim=(0, 1), ylim=(0, 1))
    ax.set_aspect('equal', 'box')
    for i in range(len(r)):
        row=r[i]
        if row[0]==0 or row[1]==0:
            continue
        if i in covered:
            ax.plot([c[i]/row[0], 0],[0, c[i]/row[1]], alpha=0.3, c='blue')
        else:
            ax.plot([c[i]/row[0], 0],[0, c[i]/row[1]], alpha=0.3, c='orange')
    pdtT=pdt.T
    ax.scatter(pdtT[0], pdtT[1], c='grey',s=10)
    ax.scatter(pdt[pdt_p, 0], pdt[pdt_p, 1], c='blue', s=10)
    ax.scatter(pdt[pdt_c, 0], pdt[pdt_c, 1], c='r', s=30)
    plt.show()
def plot_pdt_inter_hs(r, c, cost, B, inter=set()):
    fig, ax=plt.subplots()
    ax.set(xlim=(0, 1), ylim=(0, 1))
    ax.set_aspect('equal', 'box')
    ax.plot([B/cost[0], 0],[0, B/cost[1]], c='black')
    for i in range(len(r)):
        row=r[i]
        if row[0]==0 or row[1]==0:
            continue
        if i in inter:
            ax.plot([c[i]/row[0], 0],[0, c[i]/row[1]], alpha=0.3, c='red')
        else:
            ax.plot([c[i]/row[0], 0],[0, c[i]/row[1]], alpha=0.3, c='orange')
    plt.show()
def plot_inters(inter_cnt):
    fig, ax=plt.subplots()
    tmp=np.array(inter_cnt).T
    x=[1+i/10 for i in range(10)]
    for i in range(tmp.shape[0]):
        ax.plot(x,tmp[i])
    ax.set_yscale('log')
    plt.show()
```

```python
def get_sk(user, pdt, k):
    pdtT=pdt.T
    sk=np.zeros(len(user))
    for uid in range(len(user)):
        scores = np.dot(user[uid],pdtT)
        scores = sorted(scores, reverse=True)
        sk[uid]=scores[k]
    return sk
def sampleP(cardD, cardP):
    return r.sample(range(cardD), cardP)
```
